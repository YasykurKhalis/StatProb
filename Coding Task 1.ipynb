```python
import pandas as pd
import numpy as np
from scipy import stats

file_path = r'C:\Users\Yazid Arifuddin\2. Tugas Statprob\Dataset.xlsx'
df = pd.read_excel(file_path)

columns = ['Curah Hujan (mm)', 'Penjualan A (pcs)', 'Penjualan B (pcs)']

for col in columns:
    print(f"\nAnalisis ukuran pemusatan dan persebaran {col}:")
    
    mean_value = df[col].mean()

    median_value = df[col].median()

    mode_value = df[col].mode()[0]

    range_value = df[col].max() - df[col].min()

    IQR_value = df[col].quantile(0.75) - df[col].quantile(0.25)

    variance_value = df[col].var()

    std_dev_value = df[col].std()

    skewness_value = df[col].skew()

    kurtosis_value = df[col].kurtosis()

    print(f"Mean: {mean_value}")
    print(f"Median: {median_value}")
    print(f"Mode: {mode_value}")
    print(f"Range: {range_value}")
    print(f"IQR: {IQR_value}")
    print(f"Variance: {variance_value}")
    print(f"Standard Deviation: {std_dev_value}")
    print(f"Skewness: {skewness_value}")
    print(f"Kurtosis: {kurtosis_value}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

def frequency_distribution(column):
    frequency = column.value_counts().sort_index()
    frequency_table = pd.DataFrame({
        'Value': frequency.index,
        'Frequency': frequency.values
    })
    return frequency_table

for col in columns:
    print(f"\nFrequency Distribution for {col}:")
    freq_table = frequency_distribution(df[col])
    print(freq_table)

    plt.figure(figsize=(12, 6))
    plt.subplot(1, 3, 1)
    sns.histplot(df[col], bins=10, kde=True)
    plt.title(f'Histogram of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')

    plt.subplot(1, 3, 2)
    sns.kdeplot(df[col], color='blue')
    plt.title(f'Frequency Polygon of {col}')
    plt.xlabel(col)
    plt.ylabel('Density')

    plt.subplot(1, 3, 3)
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot of {col}')
    plt.xlabel(col)

    plt.tight_layout()
    plt.show()
```

```python
"""
=================================
Penjelasan tentang Fungsi Boxplot
=================================

Fungsi Boxplot:
1. Visualisasi Distribusi Data: Boxplot memberikan gambaran tentang distribusi data, menunjukkan median, kuartil, dan nilai ekstrem (outliers).
2. Identifikasi Outliers: Boxplot membantu dalam mengidentifikasi data yang berada di luar kisaran normal (outliers) dengan menggambarkan nilai ekstrem di luar whiskers.
3. Perbandingan Antara Grup: Boxplot sangat berguna untuk membandingkan distribusi data antara beberapa grup atau kategori.

======================================
Perbedaan antara Boxplot dan Histogram
======================================

Representasi Visual:
Boxplot: Menyajikan informasi tentang median, kuartil, dan outliers. Boxplot lebih fokus pada ringkasan statistik dan memungkinkan untuk membandingkan beberapa distribusi data dengan cepat.
Histogram: Menunjukkan frekuensi data dalam interval tertentu (bin). Histogram lebih baik untuk memahami bentuk distribusi dan kepadatan data.

Informasi yang Disampaikan:
Boxplot: Menyediakan informasi ringkas tentang pusat (median), sebaran (kuartil), dan pencilan (outliers) dari data.
Histogram: Menyajikan distribusi frekuensi secara lebih terperinci, memberikan pandangan visual tentang seberapa sering setiap nilai muncul.

Penggunaan:
Boxplot: Lebih sering digunakan untuk analisis ringkasan dan perbandingan antar kelompok.
Histogram: Lebih baik untuk eksplorasi data awal untuk memahami distribusi dan pola.
"""
```

```python
def identify_outliers(column):
    Q1 = column.quantile(0.25)
    Q3 = column.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = column[(column < lower_bound) | (column > upper_bound)]
    return outliers

outlier_summary = {}

for col in columns:
    outliers = identify_outliers(df[col])
    outlier_summary[col] = outliers

for col, outliers in outlier_summary.items():
    print(f"\nOutliers in {col}:")
    print(outliers)
    print(f"Number of outliers: {len(outliers)}")

total_outliers = sum(len(outliers) for outliers in outlier_summary.values())
print(f"\nTotal outliers across all columns: {total_outliers}")
```
